```python
import numpy as np
import statistics as st

# Lists of Fahrenheit temps

week_1 = [45, 55, 49, 60, 52, 55, 58]
week_2 = [51, 49, 62, 50, 53, 56, 59]
week_3 = [60, 60, 61, 62, 57, 59, 58]
week_4 = [59, 63, 62, 61, 65, 66, 62]

print(f" the mean : {np.mean(week_1)}")
print(f" the median : {np.median(week_1)}")
print(f" the mode : {st.mode(week_1)}")


#helper function 1

def central_tendency(temp_list):
 mean_temp = np.mean(temp_list)
 median_temp = np.median(temp_list)
 mode_temp = st.mode(temp_list)

 resualt = print(f"the mean : {mean_temp}\nthe median : {median_temp}\nthe mode : {mode_temp}")
 return resualt

central_tendency(week_1)



# #helper function 2

def central_tendency(temp_list , verbose = True):
    mean_temp =float(np.mean(temp_list))
    median_temp =float( np.median(temp_list))
    mode_temp = float(st.mode(temp_list))


    if verbose == True :
        print( f"the mean :{mean_temp}.\n" \
               f"the median :{median_temp}.\n" \
               f"the mode :{mode_temp}")


    elif verbose == False :
     return mean_temp, median_temp, mode_temp


central_tendency(week_3, verbose=True)
central_tendency(week_3, verbose=False)
central_tendency(week_3)








```
