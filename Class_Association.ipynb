# Class Association




```python
class BankAccount:
    def __init__(self, name, int_rate, embalanceail):
        self.name = name
        self.int_rate = int_rate
        self.embalanceail = embalanceail

    def make_deposit(self, amount):
        self.embalanceail += amount
        return self

    def make_withdrawal(self, amount):
        result = self.embalanceail - amount
        if result >= 0: # تعديل بسيط ليكون السحب ممكناً لو الرصيد صفر تماماً
            self.embalanceail = result
        else:
            print("Insufficient funds, charging a $5 fee")
            self.embalanceail -= 5
        return self

    def yield_interest(self):
        if self.embalanceail > 0:
            self.embalanceail += (self.embalanceail * self.int_rate)
        return self

    def display_user_balance(self):
        print(f"Your Name: {self.name}, Your Balance: {self.embalanceail}")

class User:
    def __init__(self, name, email):
        self.name = name
        self.email = email
        # ننشئ الحساب ونربطه بالمستخدم
        self.account = BankAccount(name=self.name, int_rate=0.02, embalanceail=0)

    def make_deposit(self, amount):
        self.account.make_deposit(amount)
        return self

    def make_withdrawal(self, amount):
        self.account.make_withdrawal(amount)
        return self

    def display_user_balance(self):
        # هنا ننادي دالة البنك لتعرض الرصيد
        print(f"User: {self.name}")
        self.account.display_user_balance()
        return self

    def transfer_money(self, other_user, amount):
        # المستخدم يسحب من حسابه ويعطي المستخدم الآخر في حسابه
        self.account.make_withdrawal(amount)
        other_user.account.make_deposit(amount)
        return self

# التشغيل الصحيح
mohammed = User("Mohammed", "Mohammed@python.com")
sara = User("Sara", "sara@python.com")

# تجربة الإيداع والطباعة
mohammed.make_deposit(100).display_user_balance()

# تجربة التحويل
print("\n--- Transfering 30 from Mohammed to Sara ---")
mohammed.transfer_money(sara, 30)
mohammed.display_user_balance()
sara.display_user_balance()
```
