```python
# example of if statement (condition is True)
a = 10
b = 20
if a < b:
  print('a is less than b')
```

```python
# example of if statement (condition is False)
a = 10
b = 20
if a > b:
  print('a is not less than b')
```

```python
# example of an if and else statement
if a < b:
  print('a is less than b')
else:
  print('a is not less than b')
```

```python
# example of an if/elif/else statement
if a < b:
  print('a is less than b')
elif a == b:
  print('a is equal to b')
else:
  print('a is greater than b')
```

```python
# redefine a
a = 100
b = 20
# example of an if/elif/else statement
if a == 100:
  print('a is equal to 100')
elif a > b:
  print('a is greater than b')
elif a < b:
  print('a is less than b')
else:
  print('a is equal to b')
```

```python
# example of two if statements
if a == 100:
  print('a is equal to 100')
if a > b:
  print('a is greater than b')
elif a < b:
  print('a is less than b')
else:
  print('a is equal to b')
```

```python
age = 17
provisional = True
# nested conditions for driving eligibility
if age < 16:
  print('You are too young to drive')
elif age >= 18:
  print('You are eligible to drive without a provisional license')
else:
  if provisional == True:
    print('You are eligible to drive with your provisional license')
  else:
    print('You need a provisional license to be eligible to drive')
```

```python
age = 18
provisional = False
# nested conditions for driving eligibility
if age < 16:
  print('You are not too young drive')
elif age >= 18:
  print('You are eligible to drive without a provisional license')
else:
  if provisional == True:
    print('You are eligible to drive with your provisional license')
  else:
    print('You need a provisional license to be eligible to drive')
```

```python
# write condition to evaluate if a number is BOTH greater than 10 AND odd
num = 13
if (num > 10) & (num % 2 != 0):
  print('number is greater than 10 and odd')
```

```python
# write condition to evaluate if a number is BOTH greater than 10 AND odd
num = 4
if (num > 10) & (num % 2 != 0):
  print('number is greater than 10 and odd')
elif (num > 10):
  print('number is greater than 10 and even')
elif (num % 2 != 0):
  print('number is less than or equal to ten and odd')
else:
  print('number is less than ten and even')
```
