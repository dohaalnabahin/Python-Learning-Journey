 # نظام استقبال المرضى (Hospital Reception)

```python
class Patient_Checkin():
  def __init__(self, patient_name, age):
        self.patient_name = patient_name
        self.age = age
        self.status = "Waiting"
        self.is_emergency =  False

  def check_up(self, temperature):
     if temperature > 30 :
      self.status = "See Doctor Now"
      return self


  def assign_emergency(self):
       self.is_emergency = True
       self.status = "Emergency Unit"
       return self

  def display_info(self)  :
    print("a report")
    print(f"patient Name : {self.patient_name}  | patient Age : {self.age}")
    print(f"patient status : {self.status}")
    print(f"Is Emergency: {self.is_emergency}")


patient1 = Patient_Checkin("Omar", 25)

patient1.check_up(39).display_info()

patient2 = Patient_Checkin("Sara", 30)
patient2.assign_emergency().display_info()


```
