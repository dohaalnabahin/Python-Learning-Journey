```python
trip_1 = [100, 50, 200, 150] # dollar
trip_2 = [70, 35, 140, 105] # dinar

# helper function 1 (المحول)
def to_dinar(amount):
    # عدلت السعر لـ 0.71
    converted_amount = amount * 0.71
    return round(converted_amount, 2)   #  للتحويل من دولار لدينار


# helper function 2 (التقرير)
def budget_extremes(expenses_list, type):
    if type == 'cheapest':           # طباعة الاكبر والاصغر
        result = min(expenses_list)
        print(f"The cheapest item cost: {result} JD")
    elif type == 'expensive':
        result = max(expenses_list)
        print(f"The expensive item cost: {result} JD")


# main function (المدير)
def travel_report(expenses, is_dollar=True):

#  استلمت العملة اذا دولار بتحول لدينار واذا دينار بتخلي زي ما هو
# هيطبع اغلشي يعني اكبر قيمة واصغر شي يعني اقل قيمة
    dinar_list = []

    if is_dollar == True:
        for x in expenses:
            converted = to_dinar(x)
            dinar_list.append(converted)
    else:
        dinar_list = expenses

    budget_extremes(dinar_list, 'cheapest')
    budget_extremes(dinar_list, 'expensive')

# --- تجربة الكود ---
print("--- Report for Trip 1 (USD) ---")
travel_report(trip_1, is_dollar=True)

print("\n--- Report for Trip 2 (JD) ---")
travel_report(trip_2, is_dollar=False)
```
