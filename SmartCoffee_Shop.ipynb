# المشروع: نظام الكافيه الذكي (Smart Coffee Shop)

```python
class Smart_Coffee_Shop:

    def __init__(self, customer_name, tax_rate=0.10):
        self.customer_name = customer_name
        self.tax_rate = tax_rate
        self.items = []
        self.total_price = 0

    def add_item(self, item_name, price):
        self.items.append(item_name)
        self.total_price += price
        return self

    def apply_discount(self, discount_amount):
        self.total_price -= discount_amount
        if self.total_price < 0:
            self.total_price = 0
        return self

    def receipt(self):
        final_price = self.total_price + (self.total_price * self.tax_rate)

        print(f"--- Receipt ---")
        print(f"Customer Name: {self.customer_name}")
        print(f"Items: {self.items}")
        print(f"Total Price (with tax): {final_price}")
        return self

order1 = Smart_Coffee_Shop("Sama", 0.15)
order1.add_item("Latte", 20).add_item("Cookie", 10).apply_discount(5).receipt()
```
