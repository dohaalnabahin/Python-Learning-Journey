# "نظام إدارة متجر إلكتروني متكامل" (E-Commerce System)

```python
class Bank_Account:
    def __init__(self, name, balance):
        self.name = name
        self.balance = balance

    def make_deposit(self, amount):
        self.balance += amount
        return self

    def make_withdrawal(self, amount):
        if self.balance >= amount:
            self.balance -= amount
            print(f"Withdrawal successful! Remaining: {self.balance}")
        else:
            print("You do not have enough credit")
        return self

class Product:
    def __init__(self, p_name, price, stock):
        self.p_name = p_name
        self.price = price
        self.stock = stock

    def reduce_stock(self, quantity):
        if quantity <= self.stock:
            self.stock -= quantity
            print(f"Stock updated! Remaining {self.p_name}: {self.stock}")
        else:
            print(f"Not enough stock! Only {self.stock} left.")
        return self # حركت الـ return للخارج

    def display_product(self):
        print(f"Product: {self.p_name} | Price: {self.price} | In Stock: {self.stock}")

class user:
    def __init__(self, U_name, Email, balance=0):
        self.U_name = U_name
        self.Email = Email
        self.account = Bank_Account(U_name, balance)

    def buy_product(self, product, quantity):
        accounting = product.price * quantity
        if self.account.balance >= accounting and product.stock >= quantity:
            self.account.make_withdrawal(accounting)
            product.reduce_stock(quantity)
            print("Completed successfully ✅")
            return self
        else:
            print("The amount is insufficient or the balance is not enough. ❌")
            return self

    def display_info(self):
        print(f"User: {self.U_name} | Balance: {self.account.balance}")


iphone = Product("iPhone 15", 1000, 5)
lana = user("Lana", "lana@email.com", 3000)

lana.buy_product(iphone, 2)
lana.display_info()
```
