```python
class Book :
   def __init__(self, title, author):
        self.title = title
        self.author = author
        self.is_borrowed = False

   def mark_as_borrowed(self):
            self.is_borrowed = True

   def mark_as_returned(self):
          self.is_borrowed = False


class user:
  def __init__(self, name, ):
        self.name = name
        self.borrowed_books = []



  def borrow_book(self, book):
    if book.is_borrowed == False :
      self.borrowed_books.append(book)
      book.mark_as_borrowed()
      print(f"Success! {self.name} borrowed '{book.title}'")

    else :
     print(f"Sorry, the book '{book.title}' is already borrowed.")


  def return_book(self, book):
   if book in self.borrowed_books:
        self.borrowed_books.remove(book)
        book.mark_as_returned()
        print(f"Success! {self.name} returned '{book.title}'")

   else:
      print(f"Error: {self.name} doesn't have this book!")


  def show_my_books(self):
    if not self.borrowed_books: # إذا كانت القائمة فارغة
        print(f"Empty list! {self.name} doesn't have any books right now.")
    else:
        print(f"--- Books currently with {self.name} ---")
        for book in self.borrowed_books:
            # هنا نصل لخاصية العنوان داخل كل كائن كتاب في القائمة
            print(f"- {book.title} (Author: {book.author})")




book1 = Book("Clean Code", "Robert Martin")
book2 = Book("Python Basics", "Lana")
me = user("Lana")

me.borrow_book(book1) # هنا نربط المستخدم بالكتاب
me.borrow_book(book2)
me.show_my_books()    # هنا نرى النتيجة

me.return_book(book1) # هنا ننهي الربط
me.show_my_books()












```
