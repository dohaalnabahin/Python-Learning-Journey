# Mean


```python
 # with loop
week_1 = [45, 55, 49, 60, 52, 55, 58]
total = 0

for i in week_1 :
 total  += i

resualt = round(total / len(week_1),2)
print (resualt)
```

```python
# without loop 2

week_2 = [45, 55, 49, 60, 52, 55, 58]
print(round(sum(week_2)/ len(week_2),2))
```

```python
def calculate_mean(numbers):
  resualt = round(sum(numbers)/ len(numbers),2)
  return resualt

week_1 = [45, 55, 49, 60, 52, 55, 58]
mean = calculate_mean(week_1)


print(mean)
```

# Median


```python
week_2 = [51, 49, 62, 50, 53, 56, 59]

week_2.sort()
n = len(week_2)
if n % 2 == 0:
    median = (week_2[n//2 - 1] + week_2[n//2]) / 2
else:
    median = week_2[n//2]

print(median)
```

```python
def calculate_median(numbers):
    numbers.sort()
    n = len(numbers)

    if n % 2 == 0:
        return (numbers[n//2 - 1] + numbers[n//2]) / 2
    else:
        return numbers[n//2]

week_2 = [51, 49, 62, 50, 53, 56, 59]
median = calculate_median(week_2)
print(median)
```

```python
import statistics

data = [51, 49, 62, 50, 53, 56, 59, 51] # مثال بيانات

# 1. حساب المتوسط
mean_v = round(statistics.mean(data), 2)

# 2. حساب الوسيط
median_v = statistics.median(data)

# 3. حساب المنوال
mode_v = statistics.mode(data)

# الطباعة بشكل مرتب
print(f"--- Statistics Summary ---")
print(f"Mean (Average): {mean_v}")
print(f"Median (Middle): {median_v}")
print(f"Mode (Frequent): {mode_v}")
```
