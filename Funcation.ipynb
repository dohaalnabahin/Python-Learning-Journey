# function
هي كتلة برمجية مُسماة يمكننا تنفيذها لأداء مهمة محددة. ببساطة، الدالة هي قائمة من التعليمات التي يمكننا تشغيلها في أي وقت وبعدد مرات غير محدود. إذا وجدنا شيئًا نستخدمه بشكل متكرر، فقد يكون من الأفضل إيجاد طريقة لتبسيط العملية

Funcation is :

- لها اسم

- تستقبل مُعاملات (الأقواس مطلوبة، والمُعاملات اختيارية)

- تنفذ سلسلة من التعليمات

- تُرجع قيمة بعد ذلك (ستُرجع None إذا لم يُذكر صراحةً أمر الإرجاع)

The advantages of using functions are:

- تقليل تكرار الكود

- تقسيم المشكلات المعقدة إلى أجزاء أبسط

- تحسين وضوح الكود

- بناء الجملة

  funcation syntax :

def: "أنا بدي أصنع دالة". def (إعلان الدالة)

الاسم: "شو بدي أسميها؟". (Function Name)

الأقواس (): "شو المواد اللي محتاجيتها عشان تشتغل؟". (Parameters)

الخلاصة: حتى لو الدالة "بخيلة" وما أعطتكِ قيمة (return) ترجعيها، هي ممكن تكون "نشيطة" وتغير في البرنامج أو تطبع على الشاشة، وهذا هو اللي بنسميه Side Effect



```python
def add(a = 3, b = 5):
    # function name: 'add', parameters: a and b
    x = a + b    # process
    return x     # return value: x
print(add())





#**************************************
    # 1. دالة مع return (بترجعلك النتيجة تمسكيها بإيدك)
def add_with_return(a, b):
    return a + b

# 2. دالة بدون return (فقط بتطبع عالشاشة وما بتقدري تستخدمي الرقم)
def add_without_return(a, b):
    print(a + b)

# الآن لو جربنا نخزن النتيجة في متغير:
result = add_with_return(5, 5) # result الآن تساوي 10، بتقدري تضربيها بـ 2 مثلاً
result2 = add_without_return(5, 5) # result2 رح تكون فاضية (None) لأنها ما رجعت شي!
```

# Parameters and Arguments
نحدد مدخلات الدوال باستخدام المعاملات. يمكن أن تحتوي الدوال على أي عدد من المعاملات، بما في ذلك الصفر

We define the input of functions using parameters. Functions can have as many parameters as we need, including 0

```python
def say_hi(name):  # say_hi funcation

    print("Hi, " + name) # name called parameters
```

الآن، يمكننا استدعاء هذه الدالة عن طريق ذكر اسمها وتمرير العدد الصحيح من الوسائط:


```python
# invoking the function 3 times, passing in one argument each time
say_hi('Michael')
say_hi('Anna')
say_hi('Eli')
```

```python
def say_hi(name):           # name هنا هو Parameter (برواز فارغ)
    print("Hi, " + name)   # نستخدم الـ Parameter داخل الدالة

say_hi('Michael')
say_hi('Anna')
say_hi('Eli')             # "Doha" هنا هو الـ Argument (الصورة الحقيقية)
```

الفرق بينهما بسيط جداً، والسر في "المكان":

1. الـ Parameter (المعلمة):
المكان: موجود في سطر تعريف الدالة (def).

الوظيفة: هو "اسم مستعار" أو "مكان محجوز" ينتظر القيمة.

مثال: name في كودك.

2. الـ Argument (المعطى):
المكان: موجود في سطر استدعاء الدالة (لما تشغليها).

الوظيفة: هو "القيمة الحقيقية" اللي بنعبيها في المكان المحجوز.

مثال: "Michael" أو "Anna".

# Add funcation
add function, recall that it takes two parameters and returns the sum of the parameters.



```python
def add(a, b):
    x = a + b
    return x

# استدعاء الدالة وتخزين النتائج (خارج الدالة)
sum1 = add(4, 6)   # sum1 ستكون 10
sum2 = add(1, 4)   # sum2 ستكون 5
sum3 = sum1 + sum2 # جمعنا النتائج المخزنة (10 + 5)

# كيف نطبع الناتج؟ ببساطة هكذا:
print(sum3)
```

# Default Parameters & Named Arguments

القيم الافتراضية (Default Parameters) تعني أنكِ تضعين قيمًا جاهزة داخل الدالة، لكي تعمل حتى لو نسي الشخص الذي يناديها إعطاءها بيانات

```python
def beCheerful(name='', repeat=2):
    # set defaults when declaring the parameters
    print(f"good morning {name}\n" * repeat)

# 1. استخدام القيم الافتراضية (No arguments)
beCheerful()

# 2. إرسال قيمة للمعطى الأول فقط (name)
beCheerful("tim")

# 3. تحديد اسم المعطى صراحة (Named argument)
beCheerful(name="john")

# 4. تغيير القيمة الافتراضية للمعطى الثاني فقط
beCheerful(repeat=6)

# 5. إرسال القيمتين معاً
beCheerful(name="michael", repeat=5)

# 6. ترتيب المعطيات لا يهم إذا حددنا الأسماء (Keyword arguments)
beCheerful(repeat=3, name="kb")
```

```python
def draw_line(shape="*", length=5):
    # shape: الرمز المراد طباعته
    # length: عدد مرات التكرار
    print(shape * length)

# الاستدعاء 1: استخدام الافتراضي
draw_line()            # سيطبع: *****

# الاستدعاء 2: نغير الرمز فقط
draw_line(shape="@")   # سيطبع: @@@@@

# الاستدعاء 3: نغير كل شيء
draw_line(shape="#", length=10) # سيطبع: ##########
```

بدون معلومات: تستخدم الدالة كل "القيم الاحتياطية" (Defaults).

معلومة واحدة (بدون اسم): تذهب للقيمة الأولى، والثانية تبقى "احتياطية".

معلومة واحدة (بالاسم): تذهب للمكان الذي سميتهِ بالضبط، والباقي "احتياطي".

معلومتان (بدون أسماء): تمشي بالترتيب (الأول للأول، والثاني للثاني).

معلومتان (بالأسماء): لا يهم الترتيب، لأنكِ حددتِ "هذا لمين وهذا لمين".
